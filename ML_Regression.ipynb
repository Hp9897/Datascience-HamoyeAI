```python
import pandas as pd
import numpy as np
import seaborn as sns
from pandas import DataFrame

df= pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv")
df.head(10)
df= pd.DataFrame(df)
print(df)
df.head(10)
df.drop(columns= ['date'])
```

```python


from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()


df= pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv")
normalised_df= df.drop(columns= ["date","lights"])
target_variable= normalised_df["Appliances"]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(normalised_df, target_variable, test_size= 0.3, random_state=42)
print(x_train, x_test, y_train)
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()

linear_model.fit(x_train,y_train)
predicted_values= linear_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae= mean_absolute_error(y_test, predicted_values)
print(mae)
round(mae, 2)
```

```python
import numpy as np
rss= np.sum(np.square(y_test, predicted_values))
round(rss, 2)
```

```python
from sklearn.metrics import mean_squared_error
rmse= np.sqrt(mean_squared_error(y_test,predicted_values))
round(rmse, 3)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```
